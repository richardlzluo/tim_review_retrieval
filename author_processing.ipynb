```python
import pandas as pd

from langchain_google_vertexai import ChatVertexAI
from langchain_openai import ChatOpenAI
from pydantic import BaseModel, Field
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage, ToolMessage
from typing import Optional, Dict, Literal, List, NotRequired, Annotated, TypedDict
import json
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
articles = pd.read_csv("articles_with_text_0214.csv")
```

```python
articles['author_list'].iloc[3]
```

```python
class AuthorList(BaseModel):
    authors: List[str] = Field(description="A list of authors in full name.")
```

```python
class AuthorListList(BaseModel):
    authors_list: List[AuthorList] = Field(description="A list of lists of authors.")
```

```python
llm_gemini = ChatVertexAI(model="gemini-2.0-pro-exp-02-05", temperature=0)
llm_gpt = ChatOpenAI(model="gpt-4o", temperature=0)

author_llm_gemini = llm_gemini.with_structured_output(AuthorListList)
author_llm_gpt = llm_gpt.with_structured_output(AuthorListList)
```

```python
system_prompt = """  
Your are given a json list of string.
Each string contains all the authors of a paper.
For every string, parse all the author names and put them into a inner list of strings. The full name of an author should be in a single string. Authors from one paper should be in a inner list.
Put all the inner lists into an outer list.
Use any character as is, including any escaping. don't try to change any characters.

"""
```

```python
llm_gemini_pro = ChatVertexAI(model="gemini-2.0-pro-exp-02-05", temperature=0)
```

```python
authors = articles['author'].to_list()
authors_json = json.dumps(authors)
```

```python
result_gemini = author_llm_gemini.invoke([SystemMessage(content=system_prompt),HumanMessage(content=authors_json)])
result_gpt = author_llm_gpt.invoke([SystemMessage(content=system_prompt), HumanMessage(content=authors_json)])

```
