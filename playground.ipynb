```python
import pandas as pd
import requests
from concurrent.futures import ProcessPoolExecutor, as_completed
from tqdm.notebook import tqdm  # Use notebook-friendly tqdm
from concurrent.futures import ThreadPoolExecutor, as_completed
from tqdm.notebook import tqdm  # Use notebook-friendly progress bar
```

```python
def fetch_all_content(df, max_threads=4):
    """
    Fetch content for each URL in the DataFrame simultaneously, printing the index being processed.

    Args:
        df (pd.DataFrame): DataFrame containing a 'url' column.
        max_threads (int): Maximum number of concurrent threads.

    Returns:
        None: Updates the DataFrame by adding a 'text' column.
    """
    text_results = [None] * len(df)  # Placeholder for results

    with ThreadPoolExecutor(max_workers=max_threads) as executor:
        # Submit tasks for each URL
        futures = {executor.submit(get_content, url): idx for idx, url in enumerate(df['url'])}

        # Task monitoring
        for future in as_completed(futures):
            idx = futures[future]  # Map future back to DataFrame index
            print(f"Processing URL at index: {idx}")  # Print the index being processed
            try:
                result = future.result()
                if isinstance(result, Exception):  # Check if an error was returned
                    print(f"Error fetching {df['url'][idx]}: {result}")
                    text_results[idx] = None
                else:
                    text_results[idx] = result
            except Exception as e:  # Catch unexpected exceptions
                print(f"Unexpected error for {df['url'][idx]}: {e}")
                text_results[idx] = None

    # Add the results to the DataFrame
    df['text'] = text_results
```
